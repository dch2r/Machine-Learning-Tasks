```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
# Generate random height (in feet) and weight (in grams) data for 40 individuals
np.random.seed(42)
heights = np.random.uniform(4, 7, 40)  # Heights between 4 and 7 feet
weights = np.random.uniform(30000, 100000, 40)  # Weights between 30,000 and 100,000 grams
```

```python
# Create a DataFrame from the generated data
data = pd.DataFrame({'Height (feet)': heights, 'Weight (grams)': weights})
```

```python
# Min-Max Normalization with custom range (-10.0, 10.0)
min_value = -10.0
max_value = 10.0
scaler_minmax = MinMaxScaler(feature_range=(min_value, max_value))
data['Normalized Weight (Min-Max)'] = scaler_minmax.fit_transform(data[['Weight (grams)']])
```

```python
# Z-score Normalization
scaler_zscore = StandardScaler()
data['Normalized Weight (Z-score)'] = scaler_zscore.fit_transform(data[['Weight (grams)']])
```

```python
# Create a scatter plot to visualize the relationship between height and normalized weight
plt.figure(figsize=(10, 6))
plt.scatter(data['Height (feet)'], data['Normalized Weight (Min-Max)'], label='Min-Max Normalized Weight', alpha=0.7)
plt.scatter(data['Height (feet)'], data['Normalized Weight (Z-score)'], label='Z-score Normalized Weight', alpha=0.7)
plt.xlabel('Height (feet)')
plt.ylabel('Normalized Weight')
plt.title('Scatter Plot: Height vs. Normalized Weight')
plt.legend()
plt.grid(True)
plt.show()
```
